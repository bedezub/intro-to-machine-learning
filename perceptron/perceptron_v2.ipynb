## Step 1: Initialization

Set initial weight and threshold 


```python
initial_weight1 = 0.3
initial_weight2 = -0.1

delta_weight1 = []
delta_weight2 = []

final_weight1 = 0
final_weight2 = 0
```

```python
threshold = 0.2
biased = -1
learning_rate = 0.1
epoch = 1 
error_sum = []        # Flag to terminate epoch loop
```

## Step 2: Activation

Activate the perceptron by applying inputs and desired output

```python
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]
```

```python
desire_outputs = [0, 1, 1, 1]
actual_outputs = []
```

For this exercise, we are using Step activation

```python
def step(sum_dot_product):
    if sum_dot_product < 0:
        return 0
    else:
        return 1
```

```python
def calc_dot_product(x1, x2, w1, w2):
    return round((x1 * w1) + (x2 * w2) + (threshold * biased), 2)
```

Activating perceptron


```python
while error_sum != 0:
    print("Epoch: ", epoch, "\n")
    error_sum = 0 
    for index, _input in enumerate(inputs):
        print("--------------- Iteration ", index, "---------------")
        sum_dot_product = calc_dot_product(_input[0], _input[1], initial_weight1, initial_weight2)
        actual_output = step(sum_dot_product)
        error = desire_outputs[index] - actual_output
        error_sum += abs(error)
        print(error)
        print("Desire output: ", desire_outputs[index])
        delta_weight1 = learning_rate * _input[0] * error
        delta_weight2 = learning_rate * _input[1] * error
        final_weight1 = round(initial_weight1 + delta_weight1, 2)
        final_weight2 = round(initial_weight2 + delta_weight2, 2)

        initial_weight1 = final_weight1
        initial_weight2 = final_weight2
        print("Sum of dot product: ", sum_dot_product, "\nActual output: ", actual_output, "\nError: ", error, "\nDelta weight 1:", delta_weight1, "\nDelta weight 2:", delta_weight2)
        print("\nerror_sum:", error_sum, "\n")
        print(final_weight1, final_weight2)
    epoch += 1
```
